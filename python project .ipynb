```python
#factorial
def fact(n):
  if n== 0:
    return 1
  else:
    return n * fact(n-1)
num = int(input())
result = fact(num)
print(result)
```

```python
#check prime number
def is_prime(num):
  if num<=1:
    return False
  for i in range(2,num):
    if num % i == 0:
      return False
    return True
num = int(input())
result= is_prime(num)
print(result)
```

```python
#fibonacci
def fibonacci(n):
  if n <= 0:
    return "invalid input"
  elif n == 1:
   return 0
  elif n == 2:
    return 1
    a , b = 0, 1
    for _ in range(n-2):
      a , b = b , a+b
  return b
num= int(input("enter the number"))
result = fibonacci(num)
print(result)
```

```python
# greatest common divisor
def gcd(a,b):
   while b :
     a,b = b,a%b
   return a
num1 = int(input())
num2 = int(input())
result = gcd(num1,num2)
print(result)
```

```python
#least common multiple
def Lcm(a,b):
  def gcd(a,b):
    while b:
      a,b = b,a%b
    return a
  Lcm = (a*b)//gcd(a,b)
  return Lcm

num1=int(input())
num2=int(input())
result = Lcm(num1,num2)
print(result)
```
